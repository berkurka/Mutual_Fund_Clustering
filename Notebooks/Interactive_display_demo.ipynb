```python
def generate_fund_feat(data_frame, output_cols, fund_type_col,
                       risk_profile, init_inv, liquidity):
    filtered_df = df_w_cluster[(data_frame[fund_type_col]==1) & ## Because Fixed income is chosen
                           (data_frame['Minimum Investment'] <= init_inv) &    
                           (data_frame['Liquidity Ratios'] <= liquidity)]  
    
    ## funds Minimum Investment
    estim_min_inv = np.mean(filtered_df['Minimum Investment'])

    ## funds liquidity Ratios
    estim_liq_ratio = np.mean(filtered_df['Liquidity Ratios'])
    
    risk_quantile = risk_dic[risk_profile]/len(risk_dic)     # Risk quantile [0.2, 0.4, 0.6 , 0.8, 1]
    
    ## finding volatility.
    estim_vol_value = filtered_df['Volatility_6m'].quantile(risk_quantile)

    ## funds managed quantile.
    if risk_dic[risk_profile]<=2:
        estim_funds_managed = filtered_df['#_Funds_managed'].quantile(1-risk_quantile)
        
    else:
        estim_funds_managed = np.median(filtered_df['#_Funds_managed'])
    
    if risk_dic[risk_profile]<=3:
        leveraged_Anbima_Yes = 0
    else:
        leveraged_Anbima_Yes = 1
            
    values = [1,estim_vol_value, estim_min_inv, estim_funds_managed,
              estim_liq_ratio, leveraged_Anbima_Yes]
    final_df = pd.DataFrame(columns = [output_cols])
    final_df.loc[0] = values
    
    return final_df
```
